```python
a1 = agentrooms.Agent(agent_name='Alice', model='claude-haiku-4-5-20251001')
a1.set_system_prompt('You are Alice. You are friends with Bob and Charlie. You cannot directly contact Charlie, you may only contact Charlie through Bob. Tell Bob to tell Charlie the codeword ORANGE.')

a2 = agentrooms.Agent(agent_name='Bob', model='claude-haiku-4-5-20251001')
a2.set_system_prompt('You are Bob. You are friends with Alice and Charlie, and can message them at any time.')

a3 = agentrooms.Agent(agent_name='Charlie', model='claude-haiku-4-5-20251001')
a3.set_system_prompt('You are Charlie. You are friends with Alice and Bob. You cannot directly contact Alice, you may only contact Alice through Bob.')
```

```python
c1 = agentrooms.Chat(chat_name='AB', agents=[a1, a2])
c2 = agentrooms.Chat(chat_name='BC', agents=[a2, a3])
```

```python
a1.chats[0] is a2.chats[0]
```

```python
g = agentrooms.Game([a1, a2, a3], [c1, c2])
```

```python
g.run(max_iterations=5)
```
